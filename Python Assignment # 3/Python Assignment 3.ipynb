```python
# 5 inputs from user for different subject’s marks, total it and generate mark sheet using grades

english = int(input("Enter Obtained Marks : "));
islamiat = int(input("Enter Obtained Marks : "));
maths = int(input("Enter Obtained Marks : "));
science = int(input("Enter Obtained Marks : "));
urdu = int(input("Enter Obtained Marks : "));
tmarks = 500;
print("Total Marks: ", tmarks);
omarks = english +  islamiat + maths + science + urdu;
print("Obtained Marks: ", omarks);
percentage = omarks/tmarks*100;
print("Percentage:",percentage);
if percentage <= 100 and percentage >=  80:
    print("Grade A+");
elif percentage <= 80  and percentage >=  70:
    print("Grade A");
elif percentage <= 70  and percentage >=  60:
    print("Grade B");
elif percentage <= 60  and percentage >=  50:
    print("Grade C");
elif percentage <= 50  and percentage >=  40:
    print("Grade D");
elif percentage <= 40  and percentage >=  33:
    print("Grade E");
elif percentage < 33  and percentage >  0:
    print("FAILED");
else:
    print("Error In Percentage");
```

```python
# sum all the numeric items in a list

def sum_list(items):
    sum_numbers = 0
    for a in items:
        sum_numbers += a
    return sum_numbers
print(sum_list([5,8,15]));
```
